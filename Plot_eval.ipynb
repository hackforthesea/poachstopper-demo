```python
import csv
import matplotlib.pyplot as plt

with open("eval_87008c9.log") as f:
    rows = list(csv.reader(f))

# dedup
results = {}
for row in rows:
    results[row[0]] = float(row[1])

results = sorted(results.items())

plt.plot([x[1] for x in results])
plt.show()
```
